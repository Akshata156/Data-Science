```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Input
```

```python
# Load the dataset
df=pd.read_csv(r"C:\Users\Akshata\OneDrive\Desktop\Assignment\Neural networks\Alphabets_data.csv")
df.head()
```

```python
# Check for missing values
print("\nMissing values in each column:")
print(df.isnull().sum())
```

```python

# Separate features and target
features = df.drop(columns=['letter'])
target = df['letter']
```

```python
# Normalize the feature columns using MinMaxScaler
scaler = MinMaxScaler()
scaled_features = scaler.fit_transform(features)
scaler
```

```python
# Convert scaled features back to a DataFrame
scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df
```

```python
# Encode the target labels (letters) into numerical values
label_encoder = LabelEncoder()
encoded_target = label_encoder.fit_transform(target)
label_encoder
```

```python
encoded_target
```

```python
# Split the dataset into training and test sets (80% training, 20% testing)
X_train, X_test, y_train, y_test = train_test_split(scaled_features, encoded_target, test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
# Model Implementation - Basic ANN
# Initialize a basic ANN model with one hidden layer
model = Sequential()
model.add(Input(shape=(X_train.shape[1],)))  # Explicit Input layer
model.add(Dense(64, activation='relu'))  # First hidden layer
model.add(Dense(32, activation='relu'))  # Another hidden layer (optional)
model.add(Dense(len(set(encoded_target)), activation='softmax'))  # Output layer
```

```python
# Compile the model
com=model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
com
```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test))
history
```

```python
# Evaluate the model
y_pred = model.predict(X_test)
y_pred_classes = y_pred.argmax(axis=1)
```

```python
# Metrics
print("\nAccuracy:", accuracy_score(y_test, y_pred_classes))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_classes))
```

```python
# Save the model architecture and weights (optional)
save=model.save("alphabet_classification_model.keras")
save
```
